```python
class ExperimentalModel:
    def __init__(self,
                 dataset,
                 model,
                 trainingSamples = 1000,
                 testSamples = 200,
                 random_seed = 4622
                 
                ):
        
        self.dataset = dataset
        
        #Create training and test data. Subsample dataset to reduce runtime.
        self.xTrain, self.xTest, self.yTrain, self.yTest = ms.train_test_split(dataset.data,
                                                                               dataset.target,
                                                                               train_size = trainingSamples,
                                                                               test_size = testSamples,
                                                                               random_state = random_seed,
                                                                               shuffle = True)

        model.fit(self.xTrain, self.yTrain)
        self.model = model
        self.training_samples = trainingSamples
        self.test_samples = testSamples
    
    def predict(self, report=True):
        
        #Generate predictions from model. Check agaisnt training error
        yhatTrain = self.model.predict(self.xTrain)
        yhatTest = self.model.predict(self.xTest)
        yhatFull = self.model.predict(self.dataset.data)
        
        
        full_balancedAcc = metrics.balanced_accuracy_score(self.dataset.target, yhatFull)
        full_f1 = metrics.f1_score(self.dataset.target, yhatFull, average = 'weighted')
        if report:
            '''
            Scoring results from model:
            - Balanced Accuracy: Useful metric when samples among populations are unbalanced
            - F1 Score: Harmonic mean between precision and accuracy.
            '''
            train_balancedAcc = metrics.balanced_accuracy_score(self.yTrain, yhatTrain)
            train_f1 = metrics.f1_score(self.yTrain, yhatTrain, average = 'weighted')

            test_balancedAcc = metrics.balanced_accuracy_score(self.yTest, yhatTest)
            test_f1 = metrics.f1_score(self.yTest, yhatTest, average = 'weighted')

            print("Model with %4d training samples and %4d testing samples" % (self.training_samples, self.test_samples))
            print("Training Scores:",
                  "\nBalanced Accuracy:\t", np.round(train_balancedAcc, 4),
                  "\nF1 Score:\t", np.round(train_f1, 4))
            print("\nTest Scores:",
                  "\nBalanced Accuracy:\t", np.round(test_balancedAcc, 4),
                  "\nF1 Score:\t", np.round(test_f1, 4))
            print("\nFull Data Scores:",
                  "\nBalanced Accuracy:\t", np.round(full_balancedAcc, 4),
                  "\nF1 Score:\t", np.round(full_f1, 4))
            
        return full_f1
```
